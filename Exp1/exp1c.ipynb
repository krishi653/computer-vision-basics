```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
RA_text = cv2.putText(blank_img, 'TAYLOR SWIFT', (120,360) , cv2.FONT_HERSHEY_TRIPLEX , 5 , (255,255) , 20, cv2.LINE_4)
```

```python
dilation_2 = cv2.dilate(RA_text,kernel,iterations=2)
plt.imshow(dilation_2)
```

```python
dilation_3 = cv2.dilate(RA_text,kernel,iterations=3)
plt.imshow(dilation_3)
```

```python
opening = cv2.morphologyEx(RA_text, cv2.MORPH_OPEN, kernel)       #erosion then dilation
plt.imshow(opening)
```

```python
closing = cv2.morphologyEx(RA_text, cv2.MORPH_CLOSE, kernel)       #dilation then erosion
plt.imshow(closing)
```

```python
gradient = cv2.morphologyEx(RA_text, cv2.MORPH_GRADIENT, kernel)       #MORPHOLOGICAL GRADIENT
plt.imshow(gradient)
```

```python
gradient_2 = cv2.morphologyEx(RA_text, cv2.MORPH_GRADIENT, kernel, iterations=2)       #MORPHOLOGICAL GRADIENT
plt.imshow(gradient_2)
```

```python
contour_img = cv2.imread('/content/Contour_img.jpg')
contour_img_rgb = cv2.cvtColor(contour_img, cv2.COLOR_BGR2RGB)
contour_img_gray = cv2.cvtColor(contour_img_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(contour_img_gray, cmap='gray')
```

```python
morph_img = cv2.imread('/content/Morphology_img_task.JPG')
morph_img_rgb = cv2.cvtColor(morph_img, cv2.COLOR_BGR2RGB)
morph_img_gray = cv2.cvtColor(morph_img_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(morph_img_gray, cmap='gray')
```

```python
thresh, morph_bi = cv2.threshold(morph_img_gray, 128,255,cv2.THRESH_BINARY)
plt.imshow(morph_bi, cmap='gray')
```

```python
kernel = np.ones((8,8))
erosion = cv2.erode(morph_bi,kernel,iterations=2)
plt.imshow(erosion, cmap='gray')
```

```python
dilation = cv2.dilate(morph_bi,kernel,iterations=2)
plt.imshow(dilation, cmap='gray')
```

```python
opening_morph = cv2.morphologyEx(morph_bi, cv2.MORPH_OPEN, kernel)       #erosion then dilation
plt.imshow(opening_morph, cmap='gray')
```

```python
closing_morph = cv2.morphologyEx(morph_bi, cv2.MORPH_CLOSE, kernel)       #dilation then erosion
plt.imshow(closing_morph, cmap='gray')
```

```python
gradient_morph = cv2.morphologyEx(morph_bi, cv2.MORPH_GRADIENT, kernel)       #MORPHOLOGICAL GRADIENT
plt.imshow(gradient_morph, cmap='gray')
```
