```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image #Shows in RGB formaT
import cv2
```

```python
dog = cv2.imread('/content/Dog.jpg')
text = cv2.imread('/content/CV_text.png')
```

```python
plt.imshow(dog)
```

```python
dog_rgb = cv2.cvtColor(dog, cv2.COLOR_BGR2RGB)
plt.imshow(dog_rgb)
```

```python
dog_gray = cv2.cvtColor(dog, cv2.COLOR_BGR2GRAY)
plt.imshow(dog_gray, cmap='gray')
```

```python
thresh, dog_bi = cv2.threshold(dog_gray, 128,255,cv2.THRESH_BINARY)
plt.imshow(dog_bi, cmap='gray')
```

```python
thresh, dog_inv = cv2.threshold(dog_gray, 127, 255, cv2.THRESH_BINARY_INV)
plt.imshow(dog_inv, cmap='gray')
```

```python
dog_resize = cv2.resize(dog_rgb,(1500,1500))
plt.imshow(dog_resize)
```

```python
plt.imshow(text)
```

```python
text_rgb = cv2.cvtColor(text, cv2.COLOR_BGR2RGB)
plt.imshow(text_rgb)
```

```python
text_gray = cv2.cvtColor(text, cv2.COLOR_BGR2GRAY)
plt.imshow(text_gray, cmap='gray')
```

```python
thresh, text_bi = cv2.threshold(text_gray, 128,255,cv2.THRESH_BINARY)
plt.imshow(text_bi, cmap='gray')
```

```python
thresh, text_inv = cv2.threshold(text_gray, 127, 255, cv2.THRESH_BINARY_INV)
plt.imshow(text_inv, cmap='gray')
```

```python
text_resize = cv2.resize(text_rgb,(1500,1500))
plt.imshow(text_resize)
```

```python
blended = cv2.addWeighted(src1=dog_resize, alpha=0.5,src2=text_resize, beta=0.3, gamma=20)
plt.imshow(blended)
```
