```python
import cv2
import numpy as np
```

```python
img1 = cv2.imread('C:/3rd Year/5th Sem/CVR data/Exp 1-3/pic1.png')
img2 = cv2.imread('C:/3rd Year/5th Sem/CVR data/Exp 1-3/pic2.png')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
sift = cv2.SIFT_create()
```

```python
keypoints1, desriptors1 = sift.detectAndCompute(img1, None)
keypoints2, desriptors2 = sift.detectAndCompute(img2, None)
```

```python
num_keypoints1 = len(keypoints1)
num_keypoints2 = len(keypoints2)
```

```python
desriptors1.shape
```

```python
desriptors2.shape
```

```python
bf_matcher = cv2.BFMatcher_create()
matches = bf_matcher.knnMatch(desriptors1, desriptors2, k=2)
```

```python
good_matches = []
for m, n in matches:
    if m.distance < 0.2 * n.distance:
        good_matches.append(m)

if len(good_matches) > 10:
    print("Image is a match.")
else:
    print("Kuch match nahi ho raha, faliure hai tu.")
```

```python
import matplotlib.pyplot as plt
matching_result = cv2.drawMatches(img1, keypoints1, img2, keypoints2, good_matches, None)
plt.imshow(matching_result)
```
