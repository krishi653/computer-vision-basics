```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img1_bgr = cv2.imread("/content/edge_image.png")
img1_gray = cv2.cvtColor(img1_bgr, cv2.COLOR_BGR2GRAY)
plt.imshow(img1_gray, cmap='gray')
plt.axis(False)
```

```python
kernel_gx = np.array([[1,0],[0,-1]])
kernel_gy = np.array([[0,1],[-1,0]])
```

```python
gradient_x = cv2.filter2D(img1_gray, cv2.CV_32F, kernel_gx)
gradient_y = cv2.filter2D(img1_gray, cv2.CV_32F, kernel_gy)
```

```python
resultant = np.sqrt(gradient_x**2 + gradient_y**2)
print(f' max: {resultant.max()}, min: {resultant.min()}')
```

```python
plt.hist(resultant.ravel())
plt.title('Histogram of Gradient Magnitude')
plt.xlabel('Gradient Magnitude')
```

```python
threshold_value = 21
edges = (resultant > threshold_value)

plt.subplot(121)
plt.imshow(img1_gray, cmap='gray')
plt.title('Original')
plt.axis(False)

plt.subplot(122)
plt.imshow(edges, cmap='gray')
plt.title('Robert')
plt.axis(False)
```

```python
prewitt_gx = np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
prewitt_gy = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
```

```python
gradient_x1 = cv2.filter2D(img1_gray, cv2.CV_32F, prewitt_gx)
gradient_y1 = cv2.filter2D(img1_gray, cv2.CV_32F, prewitt_gy)
```

```python
magnitude = cv2.magnitude(gradient_x1, gradient_y1)
```

```python
print(f'max: {magnitude.max()}, min: {magnitude.min()}')
```

```python
plt.subplot(131)
plt.imshow(img1_gray, cmap='gray')
plt.title('Original')
plt.axis(False)

plt.subplot(132)
plt.imshow(edges, cmap='gray')
plt.title('Robert edge')
plt.axis(False)

plt.subplot(133)
plt.imshow(magnitude, cmap='gray')
plt.title('Prewitt')
plt.axis(False)
```

```python
#Import "edge_image_task.png"and "edge_image_detection" image
task = cv2.imread("/content/edge_image_task.png")
detect = cv2.imread("/content/edge_image_detection.png")
```

```python
#convert image into RGB colour channel
task_rgb = cv2.cvtColor(task, cv2.COLOR_BGR2RGB)
detect_rgb = cv2.cvtColor(detect, cv2.COLOR_BGR2RGB)
```

```python
#Convert image into Grayscale channel.
task_gray = cv2.cvtColor(task_rgb, cv2.COLOR_RGB2GRAY)
detect_gray = cv2.cvtColor(detect_rgb, cv2.COLOR_RGB2GRAY)
```

```python
#Use Roberts cross egde detector for edge detection and display image by defining proper threshold value.

kernel1_gx = np.array([[1,0],[0,-1]])
kernel1_gy = np.array([[0,1],[-1,0]])

gradient1_x = cv2.filter2D(task_gray, cv2.CV_32F, kernel1_gx)
gradient1_y = cv2.filter2D(task_gray, cv2.CV_32F, kernel1_gy)

resultant1 = np.sqrt(gradient1_x**2 + gradient1_y**2)
print(f' max: {resultant1.max()}, min: {resultant1.min()}')

threshold_value = 21
edges1 = (resultant1 > threshold_value)

plt.subplot(121)
plt.imshow(task_gray, cmap='gray')
plt.title('Original')
plt.axis(False)

plt.subplot(122)
plt.imshow(edges1, cmap='gray')
plt.title('Robert')
plt.axis(False)
```
