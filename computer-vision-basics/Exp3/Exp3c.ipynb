```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import cv2
```

```python
boy = cv2.imread("/content/boy_image.jpg")
crop = cv2.imread("/content/boy_crop_template.jpg")
```

```python
boy_rgb = cv2.cvtColor(boy, cv2.COLOR_BGR2RGB)
crop_rgb = cv2.cvtColor(crop, cv2.COLOR_BGR2RGB)
```

```python
crop.shape
```

```python
template_height, template_width, channels = crop.shape
```

```python
result1 = cv2.matchTemplate(boy, crop, cv2.TM_CCOEFF_NORMED)
plt.imshow(result1)
```

```python
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result1)
```

```python
print(min_val)
print(max_loc)
print(min_loc)
print(max_loc)
```

```python
top_left = max_loc
bottom_right = (top_left[0] + template_width, top_left[1]+template_height)
```

```python
boy_detected = cv2.rectangle(boy, top_left, bottom_right, (255,255,0), 20)
```

```python
plt.imshow(boy_detected)
```

```python
result2 = cv2.matchTemplate(boy, crop, cv2.TM_SQDIFF_NORMED)
plt.imshow(result2)
```

```python
min_val1, max_val1, min_loc1, max_loc1 = cv2.minMaxLoc(result2)
```

```python
print(min_val1)
print(max_loc1)
print(min_loc1)
print(max_loc1)
```

```python
top_left1 = min_loc1
bottom_right1 = (top_left1[0] + template_width, top_left1[1]+template_height)
```

```python
boy_detected1 = cv2.rectangle(boy, top_left1, bottom_right1, (255,150,0), 20)
```

```python
plt.imshow(boy_detected1)
```
