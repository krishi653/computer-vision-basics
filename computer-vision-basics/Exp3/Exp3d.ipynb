```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
friends = cv2.imread("/content/p7892580_b_v10_ap.jpg")
plt.imshow(friends)
```

```python
friends_rgb = cv2.cvtColor(friends, cv2.COLOR_BGR2RGB)
plt.imshow(friends_rgb)
```

```python
import dlib
```

```python
face_detector_hog = dlib.get_frontal_face_detector()
```

```python
detections = face_detector_hog(friends_rgb, 1)
```

```python
detections, len(detections)
```

```python
for face in detections:
  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle(friends_rgb, (l, t), (r, b), (0, 0, 255), 2)

plt.imshow(friends_rgb)
```
